```python
# Artigo Referência: Extração de dados (arquivo zip) via URL com Python
# https://medium.com/gabifernandesc/extra%C3%A7%C3%A3o-de-dados-arquivo-zip-via-url-com-python-82a5d3d47e9d

import pandas as pd
```

```python

# url da página do zip: https://www.anbima.com.br/pt_br/informar/ranking/fundos-de-investimento/gestores.htm

url = "https://www.anbima.com.br/data/files/8F/54/F5/A4/943B281057178A286B2BA2A8/Ranking%20de%20Gestao%20-%20202207_valor.xls"

pl_por_classe = pd.read_excel(url, sheet_name='Pag. 2 - PL por Classe', skiprows=5).fillna(0)
pl_por_classe.rename(columns={'Total\n': 'Total'}, inplace=True)
pl_por_classe
```

```python
pl_por_classe[pl_por_classe['Gestor'] == 'Subtotal']
```

```python
index_ultima_linha = pl_por_classe[pl_por_classe['Gestor'] == 'Subtotal'].index.tolist()[0]
index_ultima_linha
```

```python
pl_por_classe = pl_por_classe.iloc[:index_ultima_linha]
pl_por_classe.set_index('Ordem', drop=True, inplace=True)
pl_por_classe
```

```python
colunas = pl_por_classe.columns.tolist()[1:]
colunas
```

```python
for coluna in colunas:
    pl_por_classe[coluna] = pl_por_classe[coluna].astype(float)
```

```python
pl_por_classe.info()
```

```python
pl_por_classe[pl_por_classe['Gestor'].str.contains('Tower', case=False)]
```

```python
pl_por_classe['Acoes + Mult'] = pl_por_classe['Ações'] + pl_por_classe['Multimercados']
```

```python
filtro = pl_por_classe.sort_values(by='Acoes + Mult', ascending=False).reset_index(drop=True)
filtro
```

```python
filtro[filtro['Gestor'].str.contains('Tower', case=False)]
```
