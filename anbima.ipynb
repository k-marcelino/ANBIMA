```python
# getting url from .csv file

# Imports
import pandas as pd
from datetime import date, datetime
import requests
from bs4 import BeautifulSoup

page = requests.get(f"https://www.anbima.com.br/pt_br/informar/estatisticas/fundos-de-investimento/fi-consolidado-diario.htm")
soup = BeautifulSoup(page.content, 'html.parser')

final = soup.find(id="informacoes-tecnicas").find('a')['href'][11:]

anbima = 'https://www.anbima.com.br'
url = anbima + final
url
```

```python

### Saving url in urls.txt ###
# Reading and getting last url saved
with open("urls.txt", "r") as f:
    text = f.read()
    last_line = text.split('\n')[-1]

last_line
```

```python
# Comparing last url saved with current url, if they are different, it will save the current url
if last_line != url:
    with open("urls.txt", "a") as f:
        f.write('\n' + url)
    print(f"Url: {url} adicionada com sucesso!")
else:
    print("Url já existe no arquivo!")
```

```python
# Extracting table directly from url
#url = 'https://www.anbima.com.br/data/files/37/05/E2/25/9763281033F492286B2BA2A8/20220720.xls'
movimentacao_anbima = pd.read_excel(url, skiprows=8, usecols=[0,8,9,10,11]).iloc[:7]

# Extracting date from url
file_date =  url[-12:-4]

# Inserting new column with dataRef
movimentacao_anbima.insert(0, 'dataRef', file_date)

# Changing column names
old_columns = movimentacao_anbima.columns.tolist()
new_columns = ['dataRef', 'classeAnbima', 'movimentacaoDia', 'movimentacaoMes', 'movimentacaoAno', 'movimentacao12M']
dict_from_list = dict(zip(old_columns, new_columns))
movimentacao_anbima.rename(columns=dict_from_list, inplace=True)

movimentacao_anbima
```

```python
movimentacao_anbima.to_excel(f'{path_output}{file_date}_movimentacaoFundosAnbima.xlsx', index=False, encoding='Latin-ASCII')
```
