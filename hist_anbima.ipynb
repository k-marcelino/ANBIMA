```python
# Artigo Referência: Extração de dados (arquivo zip) via URL com Python
# https://medium.com/gabifernandesc/extra%C3%A7%C3%A3o-de-dados-arquivo-zip-via-url-com-python-82a5d3d47e9d

import wget
import pandas as pd
import requests
import zipfile
from io import BytesIO
```

```python
# url da página do zip: https://www.anbima.com.br/pt_br/informar/estatisticas/fundos-de-investimento/fi-consolidado-historico.htm

link = "https://www.anbima.com.br/data/files/2E/B3/8A/C8/477C281016078A28882BA2A8/Consolidado%20Historico%20de%20Fundos%20de%20Investimento_202207.zip"
r = requests.get(link)
r
```

```python
file_bytes = BytesIO(r.content)
file_bytes
```

```python
myzip = zipfile.ZipFile(file_bytes)
myzip
```

```python
myzip.extractall()
```

```python
file_name = '\n'.join(myzip.namelist())
file_name1 = file_name.split('\n')[0]
print(file_name1)
file_name2 = file_name.split('\n')[1]
print(file_name2)
```

```python
consolidado = pd.read_excel(file_name1, usecols=[1, 12])
consolidado.dropna(inplace=True)
consolidado
```

```python
consolidado.dropna(inplace=True)
consolidado
```
